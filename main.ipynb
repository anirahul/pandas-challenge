```python
#school summary calc
school_stu_cnt = school_data_complete["school_name"].value_counts()
school_stu_cnt

group_school_stu_cnt = school_data_complete.set_index('school_name').groupby(['school_name'])

school_typ = school_data.set_index('school_name')["type"]
total_bud = school_data.set_index('school_name')["budget"]
total_students = group_school_stu_cnt["student_id"].count()
per_stu_bud = round(school_data.set_index('school_name')["budget"]/school_data.set_index('school_name')["size"],2)
                                                                
avg_read_score = group_school_stu_cnt["reading_score"].mean()
avg_math_score = group_school_stu_cnt["math_score"].mean()
                                                             
per_pass_math = ((school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["student_id"].count()/total_students)*100)

per_pass_read = ((school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["student_id"].count()/total_students)*100)

overall_pass = (per_pass_math + per_pass_read)/2
```

```python
#report data by bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete["Spending Ranges (Per Student)"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels = group_names)

#school_data_complete.head()
```

```python
#report data by bins
school_data_bins = school_data_complete.groupby("Spending Ranges (Per Student)")

avg_math = school_data_bins["math_score"].mean()
avg_read = school_data_bins["reading_score"].mean()

pass_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("Spending Ranges (Per Student)")["student_id"].count()/school_data_bins["student_id"].count()*100
pass_read = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("Spending Ranges (Per Student)")["student_id"].count()/school_data_bins["student_id"].count()*100

overall_score_bin = (pass_math + pass_read)/2

bin_frame = pd.DataFrame({
      "Average Math Score" : avg_math,
      "Average Reading Score": avg_read,
      "% Passing Math": pass_math,
      "% Passing Reading": pass_read,
      "% Overall Passing Rate": overall_score_bin
})

bin_frame

```

```python
# scores by school size
size_bins = [0, 1000, 2000, 5000]
group_names1 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["School Size"] = pd.cut(school_data_complete["size"], size_bins, labels = group_names1)

#school_data_complete.head()
```

```python
# scores by school size
school_data_sbins = school_data_complete.groupby("School Size")
savg_math = school_data_sbins["math_score"].mean()
savg_read = school_data_sbins["reading_score"].mean()

spass_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("School Size")["student_id"].count()/school_data_sbins["student_id"].count()*100
spass_read = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("School Size")["student_id"].count()/school_data_sbins["student_id"].count()*100

soverall_score_bin = (spass_math + spass_read)/2

sbin_frame = pd.DataFrame({
      "Average Math Score" : savg_math,
      "Average Reading Score": savg_read,
      "% Passing Math": spass_math,
      "% Passing Reading": spass_read,
      "% Overall Passing Rate": soverall_score_bin
})

sbin_frame
```
